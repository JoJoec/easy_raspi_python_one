# 深度学习
百度百科对深度学习的介绍(https://baike.baidu.com/item/%E6%B7%B1%E5%BA%A6%E5%AD%A6%E4%B9%A0/3729729?fr=aladdin):

深度学习(DL, Deep Learning)是机器学习(ML, Machine Learning)领域中一个新的研究方向，它被引入机器学习使其更接近于最初的目标——人工智能(AI, Artificial Intelligence)。

深度学习是学习样本数据的内在规律和表示层次，这些学习过程中获得的信息对诸如文字，图像和声音等数据的解释有很大的帮助。它的最终目标是让机器能够像人一样具有分析学习能力，能够识别文字、图像和声音等数据。 深度学习是一个复杂的机器学习算法，在语音和图像识别方面取得的效果，远远超过先前相关技术。

深度学习在搜索技术，数据挖掘，机器学习，机器翻译，自然语言处理，多媒体学习，语音，推荐和个性化技术，以及其他相关领域都取得了很多成果。深度学习使机器模仿视听和思考等人类的活动，解决了很多复杂的模式识别难题，使得人工智能相关技术取得了很大进步。

这是百度百科对深度学习的说明，这个深度相对于之前单层、浅层神经网络(BP神经网络)而言；再从单层到BP神经网络、再到现在的深度神经网络，最大的难点就在于神经元与神经元连接权值如何调整、如何找寻大量的样本来训练大量的权值。

对于当前深度学习，其深层次网络，可以提取更高层次物体特征，更好的将物体进行分类。

当前深度学习，使用卷积神经网络算法进行计算，最大的好处在于，引入了“局部感受野”的思想和权值共享的方式，极大地降低网络结构中权值的个数；权值计算，采用卷积神经网络的误差反向传播算法来调整权值。

何为误差反向传播，就是将数据输入网络中，经过计算得出输出值，输入对应的标准输出值和网络输出的输出值之差就是误差，将误差逐层逐层向输入成传播，调整每层神经网络的权值，这就是误差发现传播。

# AI+IOT=AIOT，智能物联网
智能物联网，是在物联网基础上，由于物联网中嵌入式设备算力增强，可以将深度学习算法在嵌入式端进行，这就将物联网带入了智能时代。

在AIOT方面的工作，可以划分为3的方向:

(1)深度学习算法的研究。就是当前国家研究机构、大型企业正在做的深度卷及神经网络新模型的提出和改性。这方面工作需要投入大量的人力、物力和财力，主要国家研究机构和大型企业有能力进行的工作。最简单的，深度学习计算机都是以十万级或百万级别的。

(2)深度学习算法的本地化实现。就是参照已开源的网络模型或者你能获取到的网络模型，我们针对我们硬件的计算特点，甄选出适合我们硬件的算法，或者在当前算法上改进出硬件友好型算法。例如，适合GPU的算法，不一定适合NPU或者FPGA。

(3)将已甄选出的深度学习算法使用工具或者自行开发，在本地实现、运行。这部分就是算法工程化的工作了，例如，我们把算法使用HLS在FPGA上实现。

对于从事AIOT开发的工程师，大部分从事第二、三部分工作。

后续课程主要围绕着:理解深度卷积网络、理解zynq系列PL端(就是FPGA)硬件设计特点、甄选合适深度学习算法在zynq系列PL端，这三方面工作，进行我们的AIOT课程中，本地化深度学习算法的工作。
